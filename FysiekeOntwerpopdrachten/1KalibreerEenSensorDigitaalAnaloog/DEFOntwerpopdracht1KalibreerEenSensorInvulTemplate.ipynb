## *Opdracht 6*: Meten en fitten, Leerdoel

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

I = np.array([0,0,1,3,6,7,9,14,22,24,30,35,44])
V = np.array([1,19,134,200,306,360,440,550,580,660,710,730,780])
```

```python
# Plot hier de meting en fit een lijn door de metingen.
# Maak gebruik van wat je geleerd hebt bij het practicum.

def func(x, a, b):
    return a * x ** b

val, cov = curve_fit(func, V, I)

#De errorbars zijn bepaald met intuitie en dus niet met standaard deviatie en we keken hoeveel de waardes varieerden.
I_err = I * 0.06 + 1
V_err = V * 0.02 + 3

plt.plot(V, func(V, *val), label='Curve_fit ')
plt.xlabel('#arduino')
plt.ylabel('lichtintensitieit($lumen/m^2$)')
plt.errorbar(V, I, xerr = V_err, yerr = I_err, fmt='b.', label = 'Meetpunten')
plt.legend()
plt.grid()
```
